```python
import nltk
from nltk import  word_tokenize
from nltk.corpus import words
```

#Download NLTK resources

```python
nltk.download('punkt')
nltk.download('words')
```

#Func for morphological  analysis

```python
def morphological_analysis(text):
  tokens = word_tokenize(text)
  morph_analysis = [(words, nltk.corpus.words.words()) for words in tokens]
  return morph_analysis
```

#Func for word generation

```python
def generate_words(prefix):
  return[word for word in nltk.corpus.words.words() if word.startswith(prefix)]
```

#Example  Text

```python
example_text = "The quick brown fox jumped over the lazy dog."
```

#Perform Morphological analysis

```python
analysis_result = morphological_analysis(example_text)
print("Morphologocal Analysis")
for  word, posibilities in analysis_result:
  print(f"{word} : {generate_words(word)}")
```

#Generate words with a given  prefix

```python
prefix = "In"
generated_words = generate_words(prefix)
print("\nWords starting with '{}' : {}".format(prefix, generated_words))
```
