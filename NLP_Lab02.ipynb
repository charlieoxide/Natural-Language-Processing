```python
import spacy
```

```python
def perform_ner(text):
  #load english language model
  nlp = spacy.load("en_core_web_sm")

  #process the text
  doc = nlp(text)

  #extract named entities
  entities = [(ent.text, ent.label_) for ent in doc.ents]

  return entities
```

```python
text_example = '''Narendra Damodardas Modi (born 17 September 1950) is an Indian politician.
                  He is the current Prime Minister of India serving since 2014.
                   He was the Chief Minister of the state of Gujarat from 2001 to 2014.
                  Modi was elected as the Prime Minister of India in May 2014.'''
```

```python
ansList = perform_ner(text_example)
for ent, label in ansList:
  print(f"Entity: {ent}, Label: {label}")
```
